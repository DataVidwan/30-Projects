```python
from parrot import Parrot
import torch

# Initialize Parrot model
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=torch.cuda.is_available())
```

```python
# Input text (you can replace this with your full article or split into sentences)
article_sentences = [
    "India plans to transition 30% of its vehicles to electric by 2030.",
    "The government will install 10,000 EV charging stations across the country.",
    "Environmentalists have welcomed this move as a step towards clean air."
]

# Generate paraphrased output
rewritten_article = []

for sentence in article_sentences:
    print(f"\n🔹 Original: {sentence}")
    paraphrases = parrot.augment(input_phrase=sentence, use_gpu=torch.cuda.is_available())
    if paraphrases:
        best = paraphrases[0][0]  # Take top result
        rewritten_article.append(best)
        print(f"✅ Rewritten: {best}")
    else:
        rewritten_article.append(sentence)
        print("⚠️ Could not paraphrase, using original.")

# Combine paraphrased sentences
final_output = " ".join(rewritten_article)
print("\n🔁 Final Rewritten Article:\n")
print(final_output)
```
