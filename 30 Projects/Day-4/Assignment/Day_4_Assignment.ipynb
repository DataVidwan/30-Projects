```python
import csv

file_path = "2017PurchasePricesDec.csv"

def load_inventory():
    inventory = []
    try:
        with open(file_path, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                inventory.append(row)
    except FileNotFoundError:
        print("❌ File not found!")
    return inventory

def save_inventory(inventory, fieldnames):
    with open(file_path, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        for item in inventory:
            writer.writerow(item)

def add_item(inventory, fieldnames):
    print("\n➕ Add New Item:")
    new_item = {}
    for field in fieldnames:
        print(f"Enter {field}:")
        new_item[field] = input()
    inventory.append(new_item)
    print("✅ Item added.")
    return inventory

def delete_item(inventory):
    print("Enter any field value to delete matching item:")
    key = input().lower()
    original_len = len(inventory)
    inventory[:] = [item for item in inventory if key not in [val.lower() for val in item.values()]]
    if len(inventory) < original_len:
        print("🗑️ Item deleted.")
    else:
        print("❌ Item not found.")
    return inventory

def main():
    inventory = load_inventory()
    if not inventory:
        print("No data found.")
        return
    fieldnames = list(inventory[0].keys())

    print("🛒 Inventory Manager - Add & Delete Only")
    print("1. Add Item")
    print("2. Delete Item")
    print("Choose an option (1-2):")
    choice = input()

    if choice == '1':
        inventory = add_item(inventory, fieldnames)
    elif choice == '2':
        inventory = delete_item(inventory)
    else:
        print("⚠️ Invalid choice.")

    save_inventory(inventory, fieldnames)
    print("💾 Changes saved. Program ended.")

# Run it
main()
```
