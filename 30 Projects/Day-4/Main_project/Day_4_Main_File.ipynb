```python
import csv

file_path = "student_performance_data.csv"

# Load student records
def load_students():
    students = []
    try:
        with open(file_path, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                students.append(row)
    except FileNotFoundError:
        print("CSV file not found.")
    return students

# Save updated records
def save_students(students, fieldnames):
    with open(file_path, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        for student in students:
            writer.writerow(student)

# Add student
def add_student(students, fieldnames):
    print("\n--- Add New Student ---")
    new_student = {}
    for field in fieldnames:
        new_student[field] = input(f"Enter {field}: ")
    students.append(new_student)
    print("âœ… Student added.")
    save_students(students, fieldnames)

# Delete student
def delete_student(students, fieldnames):
    print("\n--- Delete Student ---")
    key = input("Enter ID or Name to delete student: ")
    original_length = len(students)
    students[:] = [s for s in students if key not in s.values()]
    if len(students) < original_length:
        print("ðŸ—‘ï¸ Student deleted.")
    else:
        print("âŒ Student not found.")
    save_students(students, fieldnames)

# Main (runs only once)
def main():
    students = load_students()
    if not students:
        print("No data found.")
        return

    fieldnames = list(students[0].keys())

    print("\nðŸ“˜ Choose an action:")
    print("1. Add Student")
    print("2. Delete Student")
    choice = input("Enter your choice (1 or 2): ")

    if choice == '1':
        add_student(students, fieldnames)
    elif choice == '2':
        delete_student(students, fieldnames)
    else:
        print("âš ï¸ Invalid choice.")

    print("âœ… Program finished.")

# Run once
main()
```
